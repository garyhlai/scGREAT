```python
import torch
import torch.nn as nn
import numpy as np
from typing import Union, Callable
import torchdiffeq
from torch_geometric.nn import GATConv
from torch_geometric.data import Data


def is_list_like(obj) -> bool:
    return isinstance(obj, (np.ndarray, torch.Tensor, list))


def torch_t(
    t: Union[int, float, list, torch.Tensor],
    device: Union[str, torch.device] = None,
    append_zero: bool = False,
) -> torch.Tensor:
    """
    Creates the time tensor used with Neural ODEs.

    Parameters
    ----------
    t
        The time (index or list) to convert.

    device
        Device on which to put the tensor.

    append_zero
        Whether or not a 0 should come before `t` e.g. `tensor([0, 1])` if `t=1`.

    Returns
    -------
    time_tensor
        Given `t` as a tensor.
    """
    # Make sure t is a torch Tensor
    if not is_list_like(t):
        # Just a single value, integrate [0, t]
        if append_zero:
            t = torch.tensor([0, t]).float()
        # Just a single value, integrate [t]
        else:
            t = torch.tensor(t)
    elif torch.is_tensor(t):
        pass
    # is a list, but not already a torch tensor
    else:
        if append_zero:
            t = torch.tensor([0, *t]).float()
        else:
            t = torch.tensor(t)

    # Put t on correct device
    if device is not None:
        t = t.to(device)
    return t


class ODEBlock(nn.Module):
    def __init__(
        self,
        func: nn.Module,
        method: str = "dopri5",
        rtol: float = 1e-3,
        atol: float = 1e-4,
        adjoint: bool = True,
    ):
        """Standard ODEBlock class. Can handle all types of ODE functions
        :method:str = {'euler', 'rk4', 'dopri5', 'adams'}
        """
        super().__init__()
        self.func = func
        self.method = method
        self.adjoint = adjoint
        self.atol = atol
        self.rtol = rtol

    @classmethod
    def get_torchdiffeq_solver(cls, adjoint: bool) -> Callable:
        func_str = "odeint_adjoint" if adjoint else "odeint"
        return getattr(torchdiffeq, func_str)

    def forward(
        self,
        x: torch.Tensor,
        t: torch.Tensor,
        return_whole_sequence: bool = False,
        adjoint: bool = None,
    ):
        t = torch_t(t)
        t = t.to(x.device).type_as(x)

        solver = self.get_torchdiffeq_solver(
            self.adjoint if adjoint is None else adjoint
        )
        out = solver(
            self.func, x, t, rtol=self.rtol, atol=self.atol, method=self.method
        )

        if not return_whole_sequence:
            out = out[-1]

        return out


class MeanAttentionLayer(nn.Module):
    def __init__(self, axis=1):
        super().__init__()
        self.axis = axis

    def forward(self, x):
        return torch.mean(x, axis=self.axis)


class GNN(nn.Module):
    def __init__(self, in_channels, out_channels, edge_index):
        super().__init__()
        self.edge_index = edge_index
        self.conv = GATConv(
            in_channels=in_channels,
            out_channels=out_channels,
            heads=1,
            concat=True,
            dropout=0.0,
            bias=True,
        )
        self.activation = nn.Tanh()
        self.pool = MeanAttentionLayer()

    def forward(self, x):
        x = self.conv(x, self.edge_index)
        x = self.activation(x)
        x = self.pool(x)
        return x


class GDEFunc(nn.Module):
    def __init__(self, gnn: nn.Module, augment: bool = False, augment_size: int = 2):
        """General GDE function class. To be passed to an ODEBlock"""
        super().__init__()
        self.gnn = gnn

        # Number of function calls
        self.nfe = 0

        # Whether or not to augment input tensor x
        self.augment = augment
        # Dimensions of 0s to augment x with (as well as the time vector t)
        self.augment_size = augment_size

    def forward(self, t, x):
        self.nfe += 1
        if self.augment:
            x = self.augment_with_time(x, t, self.augment_size)
        # NOTE: technically dxdt
        x = self.gnn(x)
        return x

    @classmethod
    def augment_with_time(
        cls, x: torch.Tensor, t: int, num_zero_cols_to_add: int = 1
    ) -> torch.Tensor:
        """
        Augment feature matrix x with zeros and time.

        Parameters
        ----------
        x
            The input features to augment.
        t
            Time to append to x.
        num_zero_cols_to_add
            Number of columns of zeros to add to x.

        Returns
        -------
        augmented
            The augmented tensor (x, t, zeros...).
        """

        t = torch_t(t, device=x.device)
        # Augment with size number of 0s
        zeros = torch.zeros(x.size(dim=0), num_zero_cols_to_add).to(x.device)
        # Augment with time, which is only concatenated once
        times = t.repeat(x.size(dim=0), 1)
        return torch.cat((x, times, zeros), dim=1)


########### Example usage: ###########
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

######## Data ########
edge_index = torch.tensor(
    [[0, 1, 2, 0], [1, 2, 0, 2]], dtype=torch.long  # Source nodes  # Target nodes
)
in_feats = 3
out_feats = 2
num_nodes = 3
x = torch.randn(num_nodes, in_feats)


######## Hyperparameters ########
augment_size = 1
augment = True
in_channels = in_feats + 1 + augment_size

# Define GDE that models dx/dt = f(x, t), where f(x, t) is the GNN
gnn = GNN(in_channels=in_channels, out_channels=out_feats, edge_index=edge_index)
gdefunc = GDEFunc(gnn, augment=augment, augment_size=augment_size)
# `ODEBlock` is the actual model, it gets prediction by integrating the GDE
gde = ODEBlock(func=gdefunc, method="rk4", atol=1e-3, rtol=1e-4, adjoint=False).to(
    device
)

######## Forward pass ########
# Move data to the appropriate device
x = x.to(device)
edge_index = edge_index.to(device)
gde.func.gnn.edge_index = edge_index  # Ensure edge_index is on the correct device

# Define the time interval for ODE integration
t = torch.tensor([0.053, 0.753], device=device)  # From time 0 to 1
out = gde(x, t)

print("Output:", out)
```
