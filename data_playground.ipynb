```python
######################## BUILD PRIOR GRAPH FROM CORRELATION DATA ########################
import pandas as pd
import numpy as np
import random
from typing import Tuple


def construct_prior_graph_from_corr(correlation_matrix: pd.DataFrame, threshold: float) -> torch.Tensor:
    """
    Constructs a prior graph from a correlation matrix based on a given threshold.

    Parameters:
    - correlation_matrix (pd.DataFrame): The correlation matrix of gene expressions.
    - threshold (float): The correlation threshold to determine significant edges.

    Returns:
    - torch.Tensor: Edge index tensor where each column represents an edge.
    """
    # Get indices of the upper triangle of the matrix, excluding the diagonal
    upper_tri_indices = np.triu_indices_from(correlation_matrix, k=1)
    correlations = correlation_matrix.values[upper_tri_indices]
    gene_indices = list(zip(upper_tri_indices[0], upper_tri_indices[1]))

    # Filter gene pairs where absolute correlation exceeds the threshold
    significant_pairs = [(i, j) for (i, j), corr in zip(gene_indices, correlations) if abs(corr) >= threshold]

    # Convert to PyTorch tensor
    return torch.tensor(significant_pairs, dtype=torch.long).t().contiguous()

def sample_edges(num_genes: int, pos_edge_index: torch.Tensor, num_samples: int) -> Tuple[torch.Tensor, torch.Tensor]:
    """
    Samples negative edges to balance the positive edges for training.

    Parameters:
    - num_genes (int): Total number of genes.
    - pos_edge_index (torch.Tensor): Tensor of positive edges.
    - num_samples (int): Number of negative samples to generate.

    Returns:
    - Tuple[torch.Tensor, torch.Tensor]: Tensors of positive and negative edge indices.
    """
    all_possible_edges = set((i, j) for i in range(num_genes) for j in range(num_genes) if i != j)
    all_positive_edges = set(map(tuple, pos_edge_index.t().tolist()))
    all_negative_edges = all_possible_edges - all_positive_edges
    sampled_negative_edges = random.sample(list(all_negative_edges), num_samples)

    # Convert sampled negative edges to tensor
    neg_edge_index = torch.tensor(sampled_negative_edges, dtype=torch.long).t().contiguous()

    return pos_edge_index, neg_edge_index
```

```python
import torch
import torch.nn as nn
import numpy as np
from typing import Union, Callable
import torchdiffeq
from torch_geometric.nn import GATConv
from torch_geometric.data import Data
import torch.nn.functional as F

def compute_link_loss(node_embeddings, pos_edge_index, neg_edge_index):
    # Get embeddings for source and target nodes
    pos_src = node_embeddings[pos_edge_index[0]]
    pos_tgt = node_embeddings[pos_edge_index[1]]
    neg_src = node_embeddings[neg_edge_index[0]]
    neg_tgt = node_embeddings[neg_edge_index[1]]

    # Compute scores (dot product of embeddings)
    pos_scores = (pos_src * pos_tgt).sum(dim=1)
    neg_scores = (neg_src * neg_tgt).sum(dim=1)

    # Labels for positive and negative edges
    labels = torch.cat([torch.ones(pos_scores.size(0)), torch.zeros(neg_scores.size(0))]).to(node_embeddings.device)

    # Concatenate scores
    scores = torch.cat([pos_scores, neg_scores])

    # Compute binary cross-entropy loss
    loss = F.binary_cross_entropy_with_logits(scores, labels)

    return loss


def is_list_like(obj) -> bool:
    return isinstance(obj, (np.ndarray, torch.Tensor, list))


def torch_t(
    t: Union[int, float, list, torch.Tensor],
    device: Union[str, torch.device] = None,
    append_zero: bool = False,
) -> torch.Tensor:
    """
    Creates the time tensor used with Neural ODEs.

    Parameters
    ----------
    t
        The time (index or list) to convert.

    device
        Device on which to put the tensor.

    append_zero
        Whether or not a 0 should come before `t` e.g. `tensor([0, 1])` if `t=1`.

    Returns
    -------
    time_tensor
        Given `t` as a tensor.
    """
    # Make sure t is a torch Tensor
    if not is_list_like(t):
        # Just a single value, integrate [0, t]
        if append_zero:
            t = torch.tensor([0, t]).float()
        # Just a single value, integrate [t]
        else:
            t = torch.tensor(t)
    elif torch.is_tensor(t):
        pass
    # is a list, but not already a torch tensor
    else:
        if append_zero:
            t = torch.tensor([0, *t]).float()
        else:
            t = torch.tensor(t)

    # Put t on correct device
    if device is not None:
        t = t.to(device)
    return t


class ODEBlock(nn.Module):
    def __init__(
        self,
        func: nn.Module,
        method: str = "dopri5",
        rtol: float = 1e-3,
        atol: float = 1e-4,
        adjoint: bool = True,
    ):
        """Standard ODEBlock class. Can handle all types of ODE functions
        :method:str = {'euler', 'rk4', 'dopri5', 'adams'}
        """
        super().__init__()
        self.func = func
        self.method = method
        self.adjoint = adjoint
        self.atol = atol
        self.rtol = rtol

    @classmethod
    def get_torchdiffeq_solver(cls, adjoint: bool) -> Callable:
        func_str = "odeint_adjoint" if adjoint else "odeint"
        return getattr(torchdiffeq, func_str)

    def forward(
        self,
        x: torch.Tensor,
        t: torch.Tensor,
        return_whole_sequence: bool = False,
        adjoint: bool = None,
    ):
        t = torch_t(t)
        t = t.to(x.device).type_as(x)

        solver = self.get_torchdiffeq_solver(
            self.adjoint if adjoint is None else adjoint
        )
        out = solver(
            self.func, x, t, rtol=self.rtol, atol=self.atol, method=self.method
        )

        if not return_whole_sequence:
            out = out[-1]

        return out



class GNN(nn.Module):
    def __init__(self, in_channels, out_channels, edge_index):
        super().__init__()
        self.edge_index = edge_index
        self.in_channels = in_channels
        self.conv = GATConv(
            in_channels=in_channels,
            out_channels=out_channels,
            heads=1,
            concat=True,
            dropout=0.0,
            bias=True,
        )
        self.activation = nn.Tanh()

    def forward(self, x):
        x = self.conv(x, self.edge_index)
        x = self.activation(x)
        return x


class GDEFunc(nn.Module):
    def __init__(self, gnn: nn.Module, augment: bool = False, augment_size: int = 2):
        super().__init__()
        self.gnn = gnn
        self.nfe = 0
        self.augment = augment
        self.augment_size = augment_size

    def forward(self, t, x):
        self.nfe += 1
        # x should already be of shape [910, 1]
        if self.augment:
            x = self.augment_with_time(x, t, self.augment_size)
        x = self.gnn(x)
        return x


    @classmethod
    def augment_with_time(
        cls, x: torch.Tensor, t: int, num_zero_cols_to_add: int = 1
    ) -> torch.Tensor:
        """
        Augment feature matrix x with zeros and time.

        Parameters
        ----------
        x
            The input features to augment.
        t
            Time to append to x.
        num_zero_cols_to_add
            Number of columns of zeros to add to x.

        Returns
        -------
        augmented
            The augmented tensor (x, t, zeros...).
        """

        t = torch_t(t, device=x.device)
        # Augment with size number of 0s
        zeros = torch.zeros(x.size(dim=0), num_zero_cols_to_add).to(x.device)
        # Augment with time, which is only concatenated once
        times = t.repeat(x.size(dim=0), 1)
        return torch.cat((x, times, zeros), dim=1)

class DotPredictor(nn.Module):
    def forward(self, node_embeddings, edge_index):
        src = node_embeddings[edge_index[0]]
        dst = node_embeddings[edge_index[1]]
        scores = (src * dst).sum(dim=1)
        return scores
```

```python
# Function to get node features at a given time by sampling cells around that time
def get_node_features_at_time(df, time_point, genes, window_size=0.01, num_cells=50):
    # Select cells within the window around the time_point
    df_t = df[(df['PseudoTime'] >= time_point - window_size) & (df['PseudoTime'] <= time_point + window_size)]
    if df_t.empty:
        # If no cells are found, return zeros
        return np.zeros(len(genes))
    else:
        # Randomly sample cells if more than num_cells are available
        if len(df_t) > num_cells:
            df_t = df_t.sample(n=num_cells, random_state=42)
        data = df_t[genes].values
        mean_expression = data.mean(axis=0)
        return mean_expression
```

```python
from torch.optim.lr_scheduler import StepLR
# SEED the randoms
torch.manual_seed(42)
np.random.seed(42)
random.seed(42)

# Construct prior graph from correlation matrix
correlation_matrix = df_time_series.corr(method='pearson')
edge_index = construct_prior_graph_from_corr(correlation_matrix=correlation_matrix, threshold=0.7)
pos_edge_index, neg_edge_index = sample_edges(num_genes=df_expression.shape[1], pos_edge_index=edge_index, num_samples=edge_index.shape[1])

# Initialize the model
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
in_channels = 1  # Since we're using mean expression values per gene
out_channels = 1
gnn = GNN(in_channels=in_channels, out_channels=out_channels, edge_index=edge_index)
gdefunc = GDEFunc(gnn)
model = ODEBlock(func=gdefunc, method="dopri5", atol=1e-3, rtol=1e-4, adjoint=False).to(device)

# Training loop
epochs = 1
steps_per_epoch = 1  # Adjust as needed

lambda_link_loss = 10.0
optimizer = torch.optim.AdamW(model.parameters(), lr=0.01, weight_decay=5e-4)
scheduler = StepLR(optimizer, step_size=50, gamma=0.1)
criterion = nn.MSELoss()
pred = DotPredictor()

df_train = df_expression
genes = df_time_series.columns.tolist()
# Get the range of pseudotime values
pseudotime_min = df_train['PseudoTime'].min()
pseudotime_max = df_train['PseudoTime'].max()

for epoch in range(epochs):
    total_loss = 0.0
    # Number of iterations per epoch
    for _ in range(steps_per_epoch):
        # Randomly sample t0 and t1
        t0 = np.random.uniform(pseudotime_min, pseudotime_max - 0.05)
        delta_t = np.random.uniform(0.01, 0.05)  # Small time interval
        t1 = t0 + delta_t
        if t1 > pseudotime_max:
            t1 = pseudotime_max

        # Get data at time t0 and t1
        data_t0 = get_node_features_at_time(df_train, t0, genes, window_size=0.01, num_cells=50)
        data_t1 = get_node_features_at_time(df_train, t1, genes, window_size=0.01, num_cells=50)
        data_t0 = torch.Tensor(data_t0).unsqueeze(1).to(device)  # Shape: [num_genes, 1]
        data_t1 = torch.Tensor(data_t1).unsqueeze(1).to(device)
        # Skip iteration if data is empty
        if torch.all(data_t0 == 0) or torch.all(data_t1 == 0):
            print(f"Skipping iteration due to empty data. data_t0={data_t0}, data_t1={data_t1}")
            continue
        # Forward pass
        model.train()
        t_tensor = torch.tensor([t0, t1], dtype=torch.float32).to(device)
        data_tp = model(data_t0, t=t_tensor)
        # Compute losses
        mse_loss = criterion(data_tp, data_t1)
        node_embeddings = data_tp
        pos_scores = pred(node_embeddings, pos_edge_index.to(device))
        neg_scores = pred(node_embeddings, neg_edge_index.to(device))
        link_loss = compute_link_loss(pos_scores, neg_scores)
        loss = mse_loss + lambda_link_loss * link_loss
        # Backpropagation
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    scheduler.step()
    avg_loss = total_loss / steps_per_epoch
    print(f"Epoch {epoch+1}, Loss: {avg_loss:.4f}")
```
